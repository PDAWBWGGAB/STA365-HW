```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import dirichlet, multinomial
import ternary

alpha_prior = np.array([2, 3, 4])
n_trials = 100

theta_prior = dirichlet.rvs(alpha_prior, size=1)[0]

sample_multinomial = multinomial.rvs(n=n_trials, p=theta_prior)

alpha_posterior = alpha_prior + sample_multinomial

plt.figure(figsize=(6, 4))
categories = ['Category 1', 'Category 2', 'Category 3']
plt.bar(categories, sample_multinomial, color='skyblue')
plt.title('Multinomial Sample')
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.show()
```

```python
import ternary

scale = 100
figure, tax = ternary.figure(scale=scale)

tax.boundary()
tax.gridlines(multiple=10, color="blue", linewidth=0.5)
tax.left_corner_label("Category 1")
tax.right_corner_label("Category 2")
tax.top_corner_label("Category 3")

dirichlet_prior_samples = dirichlet.rvs(alpha_prior, size=500)
dirichlet_posterior_samples = dirichlet.rvs(alpha_posterior, size=500)

tax.scatter(
    dirichlet_prior_samples * scale, color="green", alpha=0.3, label="Prior"
)

tax.scatter(
    dirichlet_posterior_samples * scale, color="red", alpha=0.3, label="Posterior"
)

tax.legend()
tax.ticks(axis='lbr', linewidth=1, multiple=10)
tax.get_axes().set_title("Dirichlet Prior and Posterior")
plt.show()
```
