### What is the difference between the *acceptance probability,  proposal distribution, transition kernel, stationary distribution and target distribution*?

#### *Double Click and Answer below in this cell*

<!-- You don't have to use this answer structure template but it might help -->
**Acceptance probability**:  
  This is the factor by which we decide whether to accept a newly proposed state.  It is computed (in Metropolis–Hastings) from the target and proposal densities.

**Proposal distribution**:  
  This is the stochastic rule by which we propose a candidate next state \(\tilde{x}\) given the current state \(x^{(t)}\).  It is *not* necessarily the same as the target distribution; it can be any convenient distribution (e.g., a random walk, a discrete jump, etc.) so long as we can calculate or bound it appropriately for acceptance.

**Transition kernel**:  
  This refers to the *combined* probability of moving from \(x^{(t)}\) to \(x^{(t+1)}\).  In Metropolis–Hastings, \(K\) is made by first sampling from \(q(\tilde{x}\mid x^{(t)})\), *then* accepting or rejecting \(\tilde{x}\) according to \(\alpha_{mh}\).  Formally, \(K\) is the full conditional distribution of next-state \(\,x^{(t+1)}\) given \(x^{(t)}\).

**Stationary distribution**:  
  This is the distribution \(\pi\) (often the same as the target) such that \(\pi K = \pi\) in matrix/vector terms, or \(\pi\) remains unchanged by the Markov chain’s transition kernel \(K\).  Under typical MH conditions (aperiodic and irreducible chain), the chain converges to this stationary distribution.

**Target distribution**:  
  The distribution we *want* to sample from (for example, a posterior distribution in a Bayesian problem).  Metropolis–Hastings is designed so that \(\pi = p(\cdot)\) is a stationary distribution of the Markov chain.  Provided certain regularity conditions, the chain will converge to—and sample from—this target.

```python
# Extract waiting and eruption times
waiting = old_faithful['waiting'].values
eruptions = old_faithful['eruptions'].values
n = 272

# Model Specification
with pymc.Model() as OldFaithful:
    # 1. Beta distribution for group probability
    p = pymc.Beta("p", alpha=1, beta=1)

    # 2. Bernoulli distribution for group assignment
    u = pymc.Bernoulli("u", p=p, shape=n)

    # 3. Normal distributions for theta parameters
    theta_w0 = pymc.Normal("theta_w0", mu=55, tau=1/5**2, shape=1)
    theta_w1 = pymc.Normal("theta_w1", mu=85, tau=1/5**2, shape=1)
    theta_y0 = pymc.Normal("theta_y0", mu=2.0, tau=1/0.5**2, shape=1)
    theta_y1 = pymc.Normal("theta_y1", mu=4.5, tau=1/0.5**2, shape=1)

    # 4. Gamma distributions for tau parameters
    tau_w = pymc.Gamma("tau_w", alpha=0.5, beta=0.5*10**2, shape=1)
    tau_y = pymc.Gamma("tau_y", alpha=0.5, beta=0.5*1**2, shape=1)

    # 5. Observed Normal distributions for waiting and eruption
    obs_w = pymc.Normal(
        "obs_w",
        mu=(1-u)*theta_w0 + u*theta_w1,
        tau=tau_w,
        observed=waiting
    )

    obs_y = pymc.Normal(
        "obs_y",
        mu=(1-u)*theta_y0 + u*theta_y1,
        tau=tau_y,
        observed=eruptions
    )

    # Sampling
    trace = pymc.sample(1000, chains=4, tune=1000)

    # Posterior Analysis
    az.plot_trace(trace)
    az.summary(trace, round_to=2)
    plt.show()
```

\end{align}$

## Part 4D (10%): Complete Gibbs Sampler


```python
n = 272
G = 100

thetas_w_mu0 = np.array([55,85])
thetas_w = np.zeros((2,G)) + thetas_w_mu0.reshape(2,1)
thetas_w_tau0 = 1/5**2

thetas_y_mu0 = np.array([2.0,4.5])
thetas_y = np.zeros((2,G)) + thetas_y_mu0.reshape(2,1)
thetas_y_tau0 = 1/0.5**2

taus_alpha = 1
taus_beta = np.array([10**2, 1**2])
taus = np.zeros((2,G)) + taus_alpha/taus_beta.reshape(2,1)

us = np.zeros((n,G))
ps = np.zeros(G)+0.5

for g in range(1,G):

    # Update u
    prob_u1 = (
        stats.norm.pdf(old_faithful.waiting, loc=thetas_w[1, g-1], scale=(1/taus[0, g-1])**0.5) *
        stats.norm.pdf(old_faithful.eruptions, loc=thetas_y[1, g-1], scale=(1/taus[1, g-1])**0.5) *
        ps[g-1]
    )
    prob_u0 = (
        stats.norm.pdf(old_faithful.waiting, loc=thetas_w[0, g-1], scale=(1/taus[0, g-1])**0.5) *
        stats.norm.pdf(old_faithful.eruptions, loc=thetas_y[0, g-1], scale=(1/taus[1, g-1])**0.5) *
        (1 - ps[g-1])
    )
    prob_u = prob_u1 / (prob_u1 + prob_u0)
    us[:, g] = stats.bernoulli(prob_u).rvs()

    # Update p
    us_sum = us[:, g].sum()
    ps[g] = stats.beta(1 + us_sum, 1 + n - us_sum).rvs()

    # Update theta_w
    for k in [0, 1]:
        # Grouping data based on u
        group_data = old_faithful.waiting[us[:, g] == k]
        n_k = len(group_data)

        prec = thetas_w_tau0 + taus[0, g-1] * n_k
        tmp = thetas_w_tau0 * thetas_w_mu0[k] + taus[0, g-1] * group_data.sum()
        mu = tmp / prec

        thetas_w[k, g] = stats.norm(loc=mu, scale=1/prec**0.5).rvs()

    # Update theta_y
    for k in [0, 1]:
        # Grouping data based on u
        group_data = old_faithful.eruptions[us[:, g] == k]
        n_k = len(group_data)

        prec = thetas_y_tau0 + taus[1, g-1] * n_k
        tmp = thetas_y_tau0 * thetas_y_mu0[k] + taus[1, g-1] * group_data.sum()
        mu = tmp / prec

        thetas_y[k, g] = stats.norm(loc=mu, scale=1/prec**0.5).rvs()

    # Update tau for waiting duration
    shape_w0 = taus_alpha + len(old_faithful.waiting[us[:, g] == 0]) / 2
    rate_w0 = taus_beta[0] + 0.5 * np.sum((old_faithful.waiting[us[:, g] == 0] - thetas_w[0, g]) ** 2)
    taus[0, g] = stats.gamma(a=shape_w0, scale=1/rate_w0).rvs()

    shape_w1 = taus_alpha + len(old_faithful.waiting[us[:, g] == 1]) / 2
    rate_w1 = taus_beta[0] + 0.5 * np.sum((old_faithful.waiting[us[:, g] == 1] - thetas_w[1, g]) ** 2)
    taus[0, g] = stats.gamma(a=shape_w1, scale=1/rate_w1).rvs()

    # Update tau for eruption duration
    shape_y0 = taus_alpha + len(old_faithful.eruptions[us[:, g] == 0]) / 2
    rate_y0 = taus_beta[1] + 0.5 * np.sum((old_faithful.eruptions[us[:, g] == 0] - thetas_y[0, g]) ** 2)
    taus[1, g] = stats.gamma(a=shape_y0, scale=1/rate_y0).rvs()

    shape_y1 = taus_alpha + len(old_faithful.eruptions[us[:, g] == 1]) / 2
    rate_y1 = taus_beta[1] + 0.5 * np.sum((old_faithful.eruptions[us[:, g] == 1] - thetas_y[1, g]) ** 2)
    taus[1, g] = stats.gamma(a=shape_y1, scale=1/rate_y1).rvs()
```
